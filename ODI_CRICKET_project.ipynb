```python
# Handling missing values (Based on the distribution plot) 
# Fill missing 'age' with median  
odi1["Average"]=odi1['Average'].fillna(odi1['Average'].median(), inplace=True)
odi1["Average"]
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(y=odi1['Runs'],color="red").set_title("Boxplot for Runs")
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(y=odi1["Strike Rate"],color="yellow").set_title("Boxplot for Strike Rate")
```

```python
# Capping outliers in 'Runs' based on quantiles  
Q1 = odi1['Runs'].quantile(0.25)  
Q3 = odi1['Runs'].quantile(0.75) 
IQR = Q3 - Q1 
lower_bound = Q1 - 1.5 * IQR 
upper_bound = Q3 + 1.5 * IQR  
odi1['Runs'] = np.where(odi1['Runs'] > upper_bound, upper_bound, odi1['Runs'])  
odi1["Runs"]
```

```python
#Data transformation 
#Encode Categorical Variables  
odi1['Players'] = odi1['Players'].map({'SR Tendulkar (INDIA)': 0, 'V Kohli (INDIA)': 1, 'RT Ponting (AUS/ICC)': 2,'RG Sharma (INDIA)':3})
print(odi1.head())  
```

```python
# Data Visualization (Describe the insights)  
#Univariate Analysis 
# 1.Plotting distribution of 'Runs' 
plt.figure(figsize=(10, 6)) 
sns.histplot(odi1['Runs'], kde=True,color="green") 
plt.title('Runs Distribution') 
plt.legend(odi1["Runs"])
plt.show()  
```

```python
# 2.Box Plotting of 'Runs' 
plt.figure(figsize=(10, 6)) 
plt.boxplot(odi1['Runs']) 
plt.title('Runs Distribution') 
plt.show()  
```

```python
# 3. Scatter Plotting of 'Runs' and 'Strike rate' 
plt.figure(figsize=(10, 6)) 
plt.title("Scatter plot of Runs VS strike rate ")
plt.xlabel("Runs")
plt.ylabel("Strike rate")
plt.scatter(odi1['Runs'],odi1['Strike Rate'],color="red") 
plt.legend(labels=["Strike Rate"],title="Runs")
plt.show()  
```

```python
# Countplot for categorical features like 'Players' and 'Years of matches played' 
plt.figure(figsize=(20, 20))  
plt.subplot(1, 2, 1)  
sns.countplot(x='Players', data=odi1,color="blue")
plt.legend(labels="Players")
plt.title('Players Count')
plt.subplot(1, 2, 2) 
sns.countplot(x='Years of matches played', data=odi1,color="yellow") 
plt.legend(labels="Years of matches played")
plt.title('Years of matches Count') 
plt.show()  
```

```python
#Bivariate Analysis
# 1.Analyzing relationship between 'Players' and 'Years of matches played' 
plt.figure(figsize=(20, 20))  
sns.boxplot(x='Players', y='Years of matches played', data=odi1) 
plt.title(' Players vs Years of matches played')  
plt.show()  
```

```python
# 2.Analyzing relationship between 'Players' and 'Average' 
plt.figure(figsize=(20, 20))  
sns.boxplot(x='Players', y='Average', data=odi1) 
plt.title(' Players vs Average')  
plt.show() 
```

```python
 # 3.Analyzing relationship between '4s scored' and '6s scored' 
plt.figure(figsize=(40, 20))  
sns.boxplot(x="4s scored", y="6s scored", data=odi1,color="red") 
plt.title("4s scored vs 6s scored")  
plt.show() 
```

```python
# Multivariate Analysis
# 1.Pairplot to visualize relationships between multiple variables 
sns.pairplot(odi1[['Players', 'Years of matches played', 'Matches', 'Innings']],hue="Years of matches played") 
plt.show()  
```
